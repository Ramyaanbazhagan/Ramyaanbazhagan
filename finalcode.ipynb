```python
# =====================================
# üîß INSTALL REQUIRED LIBRARIES
# =====================================
!pip install gtts transformers sentence-transformers --quiet

# =====================================
# üß† PERSONAL AI CHATBOT
# =====================================
import datetime, random, re
from gtts import gTTS
from IPython.display import Audio, display

# Optional semantic understanding
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2')

# =====================================
# üíæ KNOWLEDGE BASE
# =====================================
facts = {
    "ramya_birthday": "29 June 2001",
    "jabez_birthday": "29 November 2001",
    "ramya_from": "Mamallapuram",
    "jabez_from": "Chrompet",
    "ramya_favourite_foods": ["mutton biryani", "mushroom pizza"],
    "ramya_dislikes": ["noodles", "dark chocolate", "brownies"],
    "jabez_job": "graphic engineer",
    "college_name": "Agni College of Technology",
    "ramya_college": "Crescent College of Engineering, Vandalur",
    "ramya_course": "M.Tech in Artificial Intelligence and Data Science",
    "memories": [
        "Ramya and Jabez went to beach on a rainy evening.",
        "Jabez gifted Ramya a ring on her birthday.",
        "They met at Agni College fest in 2019."
    ]
}

# =====================================
# üí¨ ANSWER FUNCTION
# =====================================
def chatbot_response(question):
    q = question.lower()

    # --- Pattern / keyword-based replies ---
    if "birthday" in q and "ramya" in q:
        return f"üéÇ Ramya‚Äôs birthday is on {facts['ramya_birthday']}."
    elif "birthday" in q and "jabez" in q:
        return f"üéâ Your birthday is on {facts['jabez_birthday']}."
    elif "favourite" in q and "food" in q:
        return f"üçΩÔ∏è Ramya loves {', '.join(facts['ramya_favourite_foods'])}."
    elif "dislike" in q:
        return f"üòí Ramya dislikes {', '.join(facts['ramya_dislikes'])}."
    elif "from" in q and "ramya" in q:
        return f"üèñÔ∏è Ramya is from {facts['ramya_from']}."
    elif "from" in q and "jabez" in q:
        return f"üèôÔ∏è You‚Äôre from {facts['jabez_from']}."
    elif "college" in q:
        return f"üéì Ramya studies at {facts['ramya_college']} and you studied at {facts['college_name']}."
    elif "job" in q or "work" in q:
        return f"üíº You work as a {facts['jabez_job']}."
    elif "memory" in q or "memories" in q or "moment" in q:
        return random.choice(facts["memories"])
    elif "course" in q:
        return f"üìö Ramya is pursuing {facts['ramya_course']}."

    # --- Semantic similarity fallback ---
    known = list(facts.keys())
    q_emb = model.encode(q, convert_to_tensor=True)
    known_emb = model.encode(known, convert_to_tensor=True)
    scores = util.cos_sim(q_emb, known_emb)
    best_idx = scores.argmax().item()
    if float(scores[0][best_idx]) > 0.4:
        key = known[best_idx]
        val = facts[key]
        if isinstance(val, list):
            val = ", ".join(val)
        return f"Hmm, I think you‚Äôre asking about {key.replace('_',' ')} ‚Äî it‚Äôs {val}."

    # --- Default fallback ---
    return random.choice([
        "I‚Äôm not sure yet, but I‚Äôd love to learn that!",
        "That‚Äôs interesting ‚Äî tell me more?",
        "I don‚Äôt remember that exactly, but it sounds special.",
        "Let‚Äôs find out together sometime!"
    ])

# =====================================
# üîä CHAT LOOP (run in Colab)
# =====================================
def chat():
    print("ü§ñ Personal AI is ready! Type 'exit' to stop.")
    while True:
        user = input("You: ")
        if user.lower() == "exit":
            break
        reply = chatbot_response(user)
        print("AI:", reply)

        # Speak the answer
        tts = gTTS(text=reply, lang='en')
        tts.save("reply.mp3")
        display(Audio("reply.mp3", autoplay=True))

chat()
```
